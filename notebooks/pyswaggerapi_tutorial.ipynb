```python
import os
def set_cwd_to_root(root_name: str = "PySwaggerAPIWrap") -> None:
    """
    :param root_name: Name of the root directory of the project.
    :return:

    Sets the current working directory to the root directory of the project.
    """
    new_wd = os.path.join(os.getcwd().split(root_name)[0], root_name)
    os.chdir(new_wd)

set_cwd_to_root()
```
